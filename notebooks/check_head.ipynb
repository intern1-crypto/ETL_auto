# BigQuery ロードデータの確認

`python scripts/save_preview_data.py` で保存した `preview_data/` 内の Parquet ファイルを読み込み、
各テーブルの先頭行を確認する。BigQuery への書き込みは行わない。

```python
from pathlib import Path

import pandas as pd

PREVIEW_DIR = Path("..") / "preview_data"

pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 200)
```

```python
tables = {}
for path in sorted(PREVIEW_DIR.glob("*.parquet")):
    tables[path.stem] = pd.read_parquet(path)

sorted(tables.keys())
```

```python
for name, df in tables.items():
    print(f"\n=== {name} ({df.shape[0]}行 x {df.shape[1]}列) ===")
    display(df.head())
```
